```python
 import pandas as pd
df = pd.read_csv(r"C:\Users\tvais\OneDrive\Desktop\modified_dataset.csv")
```

```python
 import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df.size
```

```python

category_counts = df['Category'].value_counts()

plt.figure(figsize=(8, 8))  
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Category Distribution")
plt.show()
```

```python
df["Type"].value_counts().head(10).plot(kind="bar",title="Top 10 Genre Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
df["Country"].value_counts().head(10).plot(kind="bar",title="Top 10 Content Producing Countries")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
df["Year"]=pd.to_datetime(df["Release_Date"]).dt.year
df["Year"].value_counts().sort_index().plot(kind="line",title="Number of Releases Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.show()
```

```python
df["Rating"].value_counts().plot(kind="bar",title="Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
df['release_month'] = pd.to_datetime(df['Release_Date']).dt.month
monthly_releases = df['release_month'].value_counts().sort_index()
monthly_releases.plot(kind='line', title="Monthly Release Frequency", figsize=(8, 6))
plt.xlabel("Month")
plt.ylabel("Number of Releases")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
top_directors=df["Director"].value_counts().head(10)
top_directors.plot(kind='bar')
plt.title("Top 10 Directors by Number of Titles")
plt.xlabel("Director")
plt.ylabel("Number of Titles")
plt.legend(["Count of Titles"], title="Legend")
plt.show()
```
